```python
# define the video path
video_path=r"video_to_image/1561.MOV"
# capture the video
cap=cv2.VideoCapture(video_path)
print(cap)
img_count=1
print(cap.isOpened())
# extract and save the video frames
while cap.isOpened():

    ret,frame= cap.read()
# This condition prevents from infinite looping
    # incase video ends.
    if not ret:
        print("unable to read frame")
        break
# Save Frame by Frame into disk using imwrite method
    print(img_count)
    if(img_count==1 or img_count%10==0):

        is_img_write=cv2.imwrite(f"video_to_image/image{img_count}.jpeg",frame)

        if is_img_write:

          print(f'image save at video_to_image/image{img_count}.jpeg')

#cv2.waitKey(25) & 0xff ==ord('q')
# increasing counter so that it will
		# show how many frames are created
    img_count+=1
# Close the window
cap.release()
# De-allocate any associated memory usage
cv2.destroyAllWindows()



```

```python
import PIL.Image
import pytesseract
from wand.image import Image as Img

import os
import cv2

image_frames='image_frames'
def files():
  try:
    os.remove(image_frames)
  except OSError:
    pass
  if not os.path.exists(image_frames):
    os.makedirs(image_frames)

  src_vid=cv2.VideoCapture("johnson.MOV")
  return(src_vid)

def process(src_vid):
  index=0
  while src_vid.isOpened():
    ret, frame = src_vid.read()
    if not ret:
      break
    name='./image_frames/frame' + str(index) +'.png'

    if index%10==0:
      print('Extracting frames....'+name)
      cv2.imwrite(name,frame)
    index=index+1
    if cv2.waitKey(10) & 0xFF == ord('q'):
      break

  src_vid.release()
  cv2.destroyAllWindows()

def get_text():
  file = open('read.txt', 'w')
  for i in os.listdir(image_frames):
    print(str(i))
    my_example=PIL.Image.open(image_frames+"/"+i)
    text=pytesseract.image_to_string(my_example, lang='eng')
    print(text)
    file.write(text)
  file.close()

if __name__=='__main__':
  vid=files()
  process(vid)
  get_text()
```
