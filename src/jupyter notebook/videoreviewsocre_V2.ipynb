```python
df=pd.read_csv("score/"+dataset+"/allscore.txt",sep = ',',names=['video_no','segment','label','senti_score','g1_score','g2_score','cloud_score'])
```

```python
df['g1_score']=df['g1_score']/df['g1_score'].max()
df['g2_score']=df['g2_score']/df['g2_score'].max()
df['cloud_score']=df['cloud_score']/df['cloud_score'].max()
df.head()
```

```python
df['average']=df[['g1_score','g2_score','cloud_score']].mean(axis=1)
df.head(23)
```

```python
df['rank']=df['average'].rank(ascending=False)
#df1=df.sort_values('rank')
#df.nsmallest(n=15,columns=['rank'])
df.head(15)
```

```python
import itertools
import pandas as pd
def jaccard_similarity(x, y):
    """ returns the jaccard similarity between two lists """
    intersection_cardinality = len(set.intersection(*[set(x), set(y)]))
    union_cardinality = len(set.union(*[set(x), set(y)]))
    return intersection_cardinality/float(union_cardinality)
final=[]
th=.7
flag=0
length=5
final_index=[]
final_index.append(0)
final.append(final_data['rank'].iloc[0])
#print(final)
#for i in final_data.index:
summarized_text=final_data['segment_sentence'].iloc[0]
for i in range(len(final_data)):
    flag=0
    for j in final_index:
        score=jaccard_similarity(final_data['segment_sentence'][i],final_data['segment_sentence'][j])
        #print(final_data['segment_sentence'][i],final_data['segment_sentence'][j])
        #print('')
        #print(score)
        #print('')
        print(final_data['rank'].loc[i],final_data['rank'].loc[j])
        if(score>th):
          flag=1
          break
    if flag==0:
      final.append(final_data['rank'].iloc[i])
      final_index.append(i)
      summer_path="score/"+dataset+"/summerized.txt"

      fh = open(summer_path, "w+")
      summaried_text=summarized_text+" "+final_data['segment_sentence'].iloc[i]
      fh.write(summarized_text)
    if len(final)==length:
      break
print(final)
#print(summarized_text)

#print(final_data['rank'][i],final_data['segment_sentence'][i])
#print(score)
```
